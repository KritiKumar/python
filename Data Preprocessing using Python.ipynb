```python
# Data Preprocessing using Python

# Importing the Data
import pandas as pd
# Load the data
automobile = pd.read_csv('Automobile.csv', sep='#', decimal='.', header=None, names=['symboling','normalized_losses','make','fuel_type','aspiration','num_of_doors','body_style','drive_wheels','engine_location','wheel_base','length','width','height','curb_weight','engine_type','num_of_cylinders','engine_size','fuel_system','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price'] )
print(automobile.dtypes)
print(automobile)

# Make 
# Check the data for typo errors and 
print(automobile['make'].value_counts())
automobile['make'] = automobile['make'].str.strip()
automobile['make'] = automobile['make'].str.lower()
automobile['make'].replace('vol00112ov', 'volvo', inplace=True)
print(automobile['make'].value_counts())

# Fuel Type
automobile['fuel_type'].value_counts()
automobile['fuel_type'] = automobile['fuel_type'].str.lower()
automobile['fuel_type'] = automobile['fuel_type'].str.strip()
print(automobile['fuel_type'].value_counts())

# Aspiration 
print(automobile['aspiration'].value_counts())
automobile['aspiration'].replace('turrrrbo', 'turbo', inplace=True)
automobile['aspiration'] = automobile['aspiration'].str.lower()
automobile['aspiration'] = automobile['aspiration'].str.strip()
print(automobile['aspiration'].value_counts())

# Number of doors
automobile['num_of_doors'].value_counts()
automobile['num_of_doors'].replace('fourR', 'four', inplace=True)
automobile['num_of_doors'] = automobile['num_of_doors'].str.lower()
automobile['num_of_doors'] = automobile['num_of_doors'].str.strip()
print(automobile['num_of_doors'].value_counts())

# Body Style
automobile['body_style'].value_counts()
automobile['body_style'] = automobile['body_style'].str.lower()
automobile['body_style'] = automobile['body_style'].str.strip()
print(automobile['body_style'].value_counts())

# Drive wheel

print(automobile['drive_wheels'].value_counts())
automobile['drive_wheels'] = automobile['drive_wheels'].str.lower()
automobile['drive_wheels'] = automobile['drive_wheels'].str.strip()
print(automobile['drive_wheels'].value_counts())

# Engine Location 

print(automobile['engine_location'].value_counts())
automobile['engine_location'] = automobile['engine_location'].str.lower()
automobile['engine_location'] = automobile['engine_location'].str.strip()
print(automobile['engine_location'].value_counts())

# Engine Type
print(automobile['engine_type'].value_counts()) 
automobile['engine_type'] = automobile['engine_type'].str.lower()
automobile['engine_type'] = automobile['engine_type'].str.strip() 
print(automobile['engine_type'].value_counts()) 

# Number of Cylinders
print(automobile['num_of_cylinders'].value_counts()) 
automobile['num_of_cylinders'] = automobile['num_of_cylinders'].str.lower()
automobile['num_of_cylinders'] = automobile['num_of_cylinders'].str.strip()
print(automobile['num_of_cylinders'].value_counts())

# Fuel System 
print(automobile['fuel_system'].value_counts()) 
automobile['fuel_system'] = automobile['fuel_system'].str.lower()
automobile['fuel_system'] = automobile['fuel_system'].str.strip()
print(automobile['fuel_system'].value_counts())

# Checking for Nulls
print(automobile['symboling'].isnull().sum())
print(automobile['normalized_losses'].isnull().sum())
print(automobile['wheel_base'].isnull().sum())
print(automobile['length'].isnull().sum())
print(automobile['width'].isnull().sum())
print(automobile['height'].isnull().sum())                
print(automobile['curb_weight'].isnull().sum())                
print(automobile['engine_size'].isnull().sum())            
print(automobile['bore'].isnull().sum()) 
print(automobile['stroke'].isnull().sum()) 
print(automobile['compression_ratio'].isnull().sum()) 
print(automobile['horsepower'].isnull().sum())                
print(automobile['peak_rpm'].isnull().sum())                
print(automobile['city_mpg'].isnull().sum())                
print(automobile['highway_mpg'].isnull().sum())                
print(automobile['price'].isnull().sum())                        

# Checking for impossible values
automobile_sym=automobile[(-3>automobile['symboling']) | (automobile['symboling'] >3)]
print(automobile_sym)
automobile_nl=automobile[(65>automobile['normalized_losses']) | (automobile['normalized_losses'] >256)]
print(automobile_nl)
automobile_wb=automobile[(86.6>automobile['wheel_base']) | (automobile['wheel_base'] >120.9)]
print(automobile_wb)
automobile_len=automobile[(141.1>automobile['length']) | (automobile['length'] >208.1)]
print(automobile_len)
automobile_wd=automobile[(60.3>automobile['width']) | (automobile['width'] >72.3)]
print(automobile_wd)
automobile_ht=automobile[(47.8>automobile['height']) | (automobile['height'] >59.8)]
print(automobile_ht)
automobile_cw=automobile[(1488>automobile['curb_weight']) | (automobile['curb_weight'] >4066)]
print(automobile_cw)
automobile_es=automobile[(61>automobile['engine_size']) | (automobile['engine_size'] >326)]
print(automobile_es)
automobile_bor=automobile[(2.54>automobile['bore']) | (automobile['bore'] >3.96)]
print(automobile_bor)
automobile_str=automobile[(2.07>automobile['stroke']) | (automobile['stroke'] >4.17)]
print(automobile_str)
automobile_cr=automobile[(7>automobile['compression_ratio']) | (automobile['compression_ratio'] >23)]
print(automobile_cr)
automobile_hp=automobile[(48>automobile['horsepower']) | (automobile['horsepower'] >288)]
print(automobile_hp)
automobile_pr=automobile[(4150>automobile['peak_rpm']) | (automobile['peak_rpm'] >6600)]
print(automobile_pr)
automobile_cmpg=automobile[(13>automobile['city_mpg']) | (automobile['city_mpg'] >49)]
print(automobile_cmpg)
automobile_hmpg=automobile[(16>automobile['highway_mpg']) | (automobile['highway_mpg'] >54)]
print(automobile_hmpg)
automobile_price=automobile[(5118>automobile['price']) | (automobile['price'] >45400)]
print(automobile_price)

# Replacing nulls with means
automobile['normalized_losses'].fillna(automobile['normalized_losses'].mean(axis=0), inplace=True)
automobile['bore'].fillna(automobile['bore'].mean(axis=0), inplace=True)
automobile['stroke'].fillna(automobile['stroke'].mean(axis=0), inplace=True)
automobile['horsepower'].fillna(automobile['horsepower'].mean(axis=0), inplace=True)
automobile['price'].fillna(automobile['price'].mean(axis=0), inplace=True)

# Replacing Outlier Values with Means and Corner Values

# Symboling
automobile['symboling'].replace(4, 3, inplace=True)
automobile_sym_withchange=automobile[(-3>automobile['symboling']) | (automobile['symboling'] >3)]
print(automobile_sym_withchange)

# Normalized Losses
mean_nl=automobile['normalized_losses'].mean(skipna=True)
print(mean_nl)
automobile['normalized_losses']=automobile['normalized_losses'].mask((65>automobile['normalized_losses']) | (automobile['normalized_losses'] >256),mean_nl)
print(automobile[(65>automobile['normalized_losses']) | (automobile['normalized_losses'] >256)])

# Price
mean_price=automobile['price'].mean(skipna=True)
print(mean_price)
automobile['price']=automobile['price'].mask((5118>automobile['price']) | (automobile['price'] >45400),mean_price)
print(automobile[(5118>automobile['price']) | (automobile['price'] >45400)])
```

```python
# Data Exploration using Python
%matplotlib inline


# Nominal Data Type : Fuel Type
import matplotlib.pyplot as plt
automobile['fuel_type'].value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
# This pie- chart supports the hypthesis that gas is the most popular fuel type as this graph helps us with frequency

# Ordinal
automobile['symboling'].plot(kind='hist',bins=20)
plt.title('Insurance Risk Rating') 
plt.show()
# The histogram shows most of the automobiles lack risk rating as majority is with 0, we do not see nay occurence with very safe risk rating which is something to think about 

# Numerical Values
automobile.boxplot(column='city_mpg')
plt.show()
# Helps detect outlier values which lie between 45 and 50, the average city mileage value comes close to 24 
```
